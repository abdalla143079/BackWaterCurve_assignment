```python
from bokeh.io import show
from bokeh.models import CustomJS, TextInput
from bokeh.io import output_file, show
from bokeh.layouts import widgetbox
from bokeh.models.widgets import RadioButtonGroup, Button
from bokeh.io import show
from bokeh.models import CustomJS, RadioButtonGroup
from bokeh.layouts import gridplot
from bokeh.layouts import column,row 
from bokeh.plotting import figure
from bokeh.models import Panel, Tabs
output_file("layout.html")
#input boxes
input_q = TextInput(value="50", title="Channel Discharge (Q m3/s)")
input_L = TextInput(value="1000", title="Channel Length (L m)")
input_B = TextInput(value="45", title="Channel Width (B m)")
input_slope = TextInput(value="0.0001", title="Bed Slope (S0)")
input_chezy= TextInput(value="50", title="Chezy Coefficient (C)")
DS_depth = TextInput(value="3", title="Water Level Dwonstream (h_d/s)")
dx_explict = TextInput(value="50", title="Computational Step (dx)")
dx_implict= TextInput(value="50", title="Computational Step (dx)")
tollerance = TextInput(value="0.001", title="Tollerance for Iteration (Error)")

#button 
btn_run = Button(label="RUN", button_type="success")
btn_close = Button(label="CLOSE", button_type="success")


btn_tabular = Button(label="Tabular results",width=200)
btn_csv = Button(label="Export csv",width=200)
btn_png = Button(label="Export png",width=200)


btns_row_1= row(btn_close,btn_run)
btns_row_0= row(btn_tabular,btn_csv,btn_png)

# figures
graph = figure(plot_width=650, plot_height=250, title=None)
graph.line([0,5,10,15,20],[0,5,10,15,20])
fig= figure(plot_width=650, plot_height=650, title=None)

####### rdio button #####
LABELS = ["Explicit", "implicit"]
solution_mthd = RadioButtonGroup(labels=LABELS)


# displaying 

basic_input = column(input_q, input_L, input_B,input_slope, input_chezy,DS_depth)
numerical_input = column(solution_mthd, dx_explict,dx_implict,tollerance)

c=column( row(basic_input,numerical_input),graph,btns_row_0, btns_row_1)


#tabs

tab_simulation = Panel(child=c, title="simulation ")
tab_BWC = Panel(child=fig,title="BWC",)
tab_implicit = Panel( child=fig,title="IMPLICIT")
tab_explict = Panel(child=fig,title="EXPLICIT")
tab_composed= Tabs(tabs=[tab_simulation, tab_BWC,tab_implicit,tab_explict])
#sfasdfadsfdsafasdf
show(Tabs(tabs=[tab_simulation,tab_BWC,tab_explict,tab_implicit],width=2000))
```
